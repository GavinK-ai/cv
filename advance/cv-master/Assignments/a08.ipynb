```python
import numpy as np

dataZeros = np.zeros((10,5,3),dtype='uint8')
#dataZero
```

```python
data255 = np.ones((10,5,3),dtype='uint8')*255
#data255
```

```python
dataHStack = np.hstack((dataZeros, data255))
dataVStack = np.vstack((dataZeros, data255))
```

```python
import matplotlib.pyplot as plt
plt.figure(figsize=(20,10))
plt.subplot(1,4,1)
plt.imshow(dataZeros)
plt.title('Array of 0')

plt.subplot(1,4,2)
plt.imshow(data255)
plt.title('Array of 255')

plt.subplot(1,4,3)
plt.imshow(dataHStack)
plt.title('Array of 0 & 255 Horizontally')

plt.subplot(1,4,4)
plt.imshow(dataVStack)
plt.title('Array of 0 & 255 Vertiaclly')
```

```python
np.random.seed = 11
randColor = np.random.randint(0,255,(50,50,3))

plt.figure(figsize=(20,10))
plt.subplot(1,4,1)
plt.imshow(randColor)
plt.title('Array of Color')
```

```python
topBar = randColor[:26, :, :]
btmBar = randColor[26:, :, :]

red = np.full((25, 50, 3), [255, 0, 0])
green = np.full((25, 50, 3), [0, 255, 0])

red_green = np.vstack((red,green))

topBar = red
btmBar = green

plt.figure(figsize=(20,10))
plt.subplot(1,2,1)
plt.imshow(randColor)
plt.title('Array of Color')

plt.subplot(1,2,2)
plt.imshow(red_green)
plt.title('Array of Red Green')
```

```python
randColor1 = np.copy(randColor)

top = randColor1[:1, :, :]
mid = randColor1[25:26, :, :]
btm = randColor1[-1:, :, :]

redBar = np.full((1, 50, 3), [255, 0, 0])
greenBar = np.full((1, 50, 3), [0, 255, 0])
blueBar = np.full((1, 50, 3), [0, 0, 255])

randColor1 = np.where(randColor == top, redBar, randColor1)
randColor1 = np.where(randColor == mid, blueBar, randColor1)
randColor1 = np.where(randColor == btm, greenBar, randColor1)

randColor2 = np.copy(randColor)
left = randColor2[:, :1, :]
cen = randColor2[:, 25:26, :]
right = randColor2[:, -1:, :]

purpleBar = np.full((50, 1, 3), [255, 0, 255])
yellowBar = np.full((50, 1, 3), [255, 255, 0])
cyanBar = np.full((50, 1, 3), [0, 255, 255])

randColor2 = np.where(randColor == left, purpleBar, randColor2)
randColor2 = np.where(randColor == cen, yellowBar, randColor2)
randColor2 = np.where(randColor == right, cyanBar, randColor2)

plt.figure(figsize=(20, 10))
plt.subplot(1, 3, 1)
plt.imshow(randColor)
plt.title('Array of Color')

plt.subplot(1, 3, 2)
plt.imshow(randColor1)
plt.title('Array of Red Green Blue Strip')

plt.subplot(1, 3, 3)
plt.imshow(randColor2)
plt.title('Array of Purple Yellow Cyan Strip')
```
