```python
# 1. Create an array of shape (100, 100, 3) of unsigned integer 8 bits filled with random values between 0 and 256.

import numpy as np

# seed the numpy random egine for constant or future replication of data
np.random.seed(12345)

# create array
a = np.random.randint(0, 256, (100, 100, 3), dtype=np.uint8)

```

```python
import matplotlib.pyplot as plt

plt.imshow(a)
plt.show()
```

```python
# show the RGB
a[0,0,:]
```

```python
# 4. Copy array from question 1 and change the values that are less than 110 to 0 and others to 255.

b = a.copy()

b[b < 110]= 0
b[b >= 110]= 255
```

```python
import matplotlib.pyplot as plt

plt.imshow(b)
plt.show()
```
