```python
# 1. Remove the blue color component from image ca(.ext) and display.
# 2. Remove the green color component from image ca(.ext) and display.
# 3. Remove the red color component from image ca(.ext) and display.

import cv2 as cv
import matplotlib.pyplot as plt

# Read the image

path1 = r'assets/rgb.png'
path2 = r'assets/ca.png'
path3 = r'assets/ca2.png'

path_ls = [path1, path2, path3]

for x in range(len(path_ls)):
    img = cv.imread(path_ls[x])

    # Convert from BGR to RGB
    img = img[:,:,::-1]

    # Copy the original image for each color component
    imgNoBlue = img.copy()
    imgNoGreen = img.copy()
    imgNoRed = img.copy()

    # Set the designated colour value to 0
    imgNoRed[:,:,0] = 0
    imgNoGreen[:,:,1] = 0
    imgNoBlue[:,:,2] = 0

    img_ls = [img, imgNoRed, imgNoGreen, imgNoBlue]

    # Initialize plot parameter
    plt.figure(figsize=(30,10))
    fsize = 25
    
    # Display the images
    for i in range(len(img_ls)):
        plt.subplot(1,4,i+1)
        plt.imshow(img_ls[i])
        plt.axis('off')

        if i == 0:
            plt.title('Original', fontsize=fsize)
        elif i == 1:
            plt.title('No Red', fontsize=fsize)
        elif i == 2:
            plt.title('No Green', fontsize=fsize)
        elif i == 3:
            plt.title('No Blue', fontsize=fsize)


plt.show()
```

```python
# 4. Apply all the edge detection to ca[.ext] and display

import cv2 as cv
import numpy as np
import matplotlib.pyplot as plt

# Read the image & convert to RGB
path = r'assets/dog.png'
img = cv.imread(path)
img = img[:,:,::-1]
    
# Define the kernels
kernel_edge_1 = np.array([[1, 0, -1],
                        [0, 0, 0], 
                        [-1, 0, 1]])

kernel_edge_2 = np.array([[0, 1, 0],
                        [1, -4, 1],
                        [0, 1, 0]])

kernel_edge_3 = np.array([[-1, -1, -1],
                        [-1, 8, -1],
                        [-1, -1, -1]])

sobel_x_kernel = np.array([[1, 0, -1],
                           [2, 0, -2],
                           [1, 0, -1]])

sobel_y_kernel = np.array([[1, 2, 1],
                           [0, 0, 0],
                           [-1, -2, -1]])


# Apply filtering based on kernel
img_kernel_edge_1 = cv.filter2D(img, -1, kernel_edge_1)
img_kernel_edge_2 = cv.filter2D(img, -1, kernel_edge_2)
img_kernel_edge_3 = cv.filter2D(img, -1, kernel_edge_3)
img_sobel_x_kernel = cv.filter2D(img, -1, sobel_x_kernel)
img_sobel_y_kernel = cv.filter2D(img, -1, sobel_y_kernel)

# Keep images in a list to display
img_ls = [img, img_kernel_edge_1, img_kernel_edge_2, img_kernel_edge_3, img_sobel_x_kernel, img_sobel_y_kernel]

# Initialization
nrows = 3
ncols = 2
fsize = 20
plt.figure(figsize=(20,20))

# Display the figure
for i in range(len(img_ls)):
    
    plt.subplot(nrows,ncols,i+1)
    plt.imshow(img_ls[i])
    plt.axis('off')
    
    if i == 0:
        plt.title('Original',fontsize=fsize)
    elif i == 1:
        plt.title('Edge Detection 1',fontsize=fsize)
    elif i == 2:
        plt.title('Edge Detection 2',fontsize=fsize)
    elif i == 3:
        plt.title('Edge Detection 3',fontsize=fsize)
    elif i == 4:
        plt.title('Edge Detection Sobel X',fontsize=fsize)
    elif i == 5:
        plt.title('Edge Detection Sobel Y',fontsize=fsize)

#plt.savefig('AssignmentEdgeDetection.jpg', bbox_inches='tight', dpi=300)
plt.show()
```
